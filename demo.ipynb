### Regression Demo

```python
import imports as im
import random
import numpy as np

# creates the traning data
training_samples = {}
def mapping(x,y,z):
    return 2*x + np.sin(y) + 4*z
for i in range(30000):
    x = random.uniform(-5,5)
    y = random.uniform(-5,5)
    z = random.uniform(-5,5)
    w = mapping(x,y,z)
    training_samples[(x,y,z)] = w

# creates the testing data
testing_samples = {}
for i in range(20):
    x = random.uniform(-5,5)
    y = random.uniform(-5,5)
    z = random.uniform(-5,5)
    w = mapping(x,y,z)
    testing_samples[(x,y,z)] = w

# customize network
network = im.Network([im.Layer(3, 4, 'sigmoid'),
                      im.Layer(4, 6, 'relu'),
                      im.Layer(6, 3, 'relu'),
                      im.Layer(3, 1, "linear")], 
                      rate=0.001,
                      regressor=True)

# tests and trains the network
# returns a dataframe of predicted and true ys
network.train_test(
                training_samples=training_samples, 
                testing_samples=testing_samples, 
                batch_size=64,
                n_epochs=15
                )
```
